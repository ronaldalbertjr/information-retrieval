```python
import numpy as np
```

```python
# Definindo a matriz de transições da rede descrita no trabalho
A = np.column_stack([[0, 1/2, 1/2, 0],[0, 0, 1, 0], [1, 0, 0, 0], [0, 0, 1, 0]])
```

```python
# Implementação do algoritmo de Pagerank
def Pagerank(matrizTransform, erroMin):
    v = np.ones((len(matrizTransform[:,1]), 1)) * (1/len(matrizTransform[:,1]))
    erro = 1
  
    while(erro > erroMin):
        if(erro > np.linalg.norm(v - matrizTransform@v)):
            erro = np.linalg.norm(v - matrizTransform@v)
        v = matrizTransform@v
  
    return v


resultado = Pagerank(A, 0.0001)
resultado
```

```python
# Definindo T = Número de páginas
T = 4

#Definindo beta para o teleporte rândomico
beta = 0.8
```

```python
# Definindo a função para aplicar o teleporte randômico
TeleporteRandômico =  lambda resultadoPagerank, beta, T: ((1-beta)/T) + beta*resultadoPagerank

TeleporteRandômico(resultado, beta, T)
```
